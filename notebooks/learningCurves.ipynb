```python
# Import modules
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read csv
df = pd.read_csv("learningCurves.txt", index_col="Number of output frames")

df
```

```python
colours = ["blue","orange","green","cyan","purple"]
frames = list(df.index)

for i in range(len(df)):
    plt.plot(df.iloc[i], label=(str(df.index[i]) + " Frames"), color=colours[i])

plt.xlabel("Epoch number")
plt.ylabel("Training Loss")
plt.title("Rolling Training Learning Curves")
plt.legend()
# plt.savefig("noRollingValidationLearningCurves.png")
```
